```python
import pandas as pd
from pathlib import Path 


data_dir = r'C:\SportsProject\premier_team_analysis\data'

player_stats_df = pd.read_csv(f'{data_dir}/player_stats.csv')
player_salaries_df = pd.read_csv(f'{data_dir}/player_salaries.csv')
team_stats_df = pd.read_csv(f'{data_dir}/team_stats.csv')
standings_df = pd.read_csv(f'{data_dir}/standings.csv')
fixtures_df = pd.read_csv(f'{data_dir}/fixtures.csv')
team_salary_df = pd.read_csv(f'{data_dir}/team_salary.csv')
player_possession_stats_df = pd.read_csv(f'{data_dir}/player_possession_stats.csv')
team_possession_stats_df = pd.read_csv(f'{data_dir}/team_possession_stats.csv')


#Code from https://www.kaggle.com/code/astonvaz/stats-don-t-lie-premier-league-24-25-deep-dive

df_dict = {
    'team_possession_stats': team_possession_stats_df,
    'player_possession_stats': player_possession_stats_df,
    'team_salary': team_salary_df,
    'standings': standings_df,
    'team_stats': team_stats_df,
    'player_salaries': player_salaries_df,
    'fixtures': fixtures_df,
    'player_stats': player_stats_df
}

def print_df_shape(name, dataframe):
    """Displays dataframe name and shape"""
    print(f"{name} Dataset shape: {dataframe.shape}\n")
    
def display_column_names(name, dataframe):
    """Displays column names in a dataframe"""
    print(f"{name} columns:\n{dataframe.columns.tolist()} \n")
    
def display_first_five_rows(table_name, dataframe):
    print(f"{table_name} data:")
    print(f"{dataframe.head(5)}\n\n")
    
def display_data_types(name, dataframe):
    print(f"{name} data types:")
    print(f"{dataframe.info()}\n")
    
    
```

```python
player_stats = df_dict['player_stats']
player_possession = df_dict['player_possession_stats']
player_salaries = df_dict['player_salaries']


# Merge on player name (normalize columns)
player_stats = player_stats.rename(columns={'name': 'player'})
player_salaries = player_salaries.rename(columns={'Player': 'player'})

# Merge all data
player_df = player_stats.merge(player_possession, on=['player', 'team', 'age'], suffixes=('', '_pos'), how='inner')
player_df = player_df.merge(player_salaries[['player', 'Annual']], on='player', how='left')

display(player_df)




```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import numpy as np

MIN_MINUTES = 300
"""
#wights to give more importance to some predictors
weights = {
    'goals_per90': 5,
    'assists_per90': 4,
    'expected_goals_per90': 4,
    'progressive_passes_per90': 3,
    'progressive_carries_per90': 3,
    'touches_per90': 1,
    'carries_per90': 1,
    'total_distance_carried_per90': 0.5,
    'received_per90': 1,
    'attempted_take_ons_per90': 2,
    'successful_take_ons_per90': 2,
}
"""


# Minimum minuts to avoid outliers

player_df = player_df[player_df['minutes'] >= MIN_MINUTES].copy()

# Extract first position (before "," or "/")
player_df['position_clean'] = player_df['position'].astype(str).str.split('[,/]').str[0].str.strip()

# Optional: ensure it falls into standard categories
valid_positions = ['GK', 'DF', 'MF', 'FW']
player_df['position_clean'] = player_df['position_clean'].apply(lambda x: x if x in valid_positions else 'Other')



# Create per-90 columns
per90_features = ['goals', 'assists', 'expected_goals', 'touches', 'carries',
                  'total_distance_carried', 'received', 'progressive_passes',
                  'progressive_carries', 'attempted_take_ons', 'successful_take_ons']

for col in per90_features:
    player_df[f'{col}_per90'] = player_df[col] / (player_df['minutes'] / 90)

# Final features to use
features = [f'{col}_per90' for col in per90_features] 



X = player_df[features].fillna(0)
# Apply weights before scaling
X_weighted = X.copy()
for col in weights:
    X_weighted[col] = X[col] * weights[col]
    
    


from sklearn.preprocessing import MinMaxScaler

#scaler = MinMaxScaler()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



kmeans = KMeans(n_clusters=4, random_state=42)
player_df['cluster'] = kmeans.fit_predict(X_scaled)



# PCA transformation
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['cluster'] = player_df['cluster']
pca_df['position'] = player_df['position_clean']
pca_df['player'] = player_df['player']

# Plot with better colors + simplified shapes
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=pca_df,
    x='PC1', y='PC2',
    hue='cluster',         # color = cluster
    style='position',      # shape = position group
    palette='tab10',       # or try 'Set2', 'Dark2', 'husl'
    s=100
)
plt.title('Player Clusters (PCA)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()



# Compute distances to each cluster center
distances = kmeans.transform(X_scaled)

# Add distance to own cluster
player_df['distance_to_center'] = distances[np.arange(len(distances)), player_df['cluster']]

# Top 3 most representative per cluster
for c in range(kmeans.n_clusters):
    print(f"\nðŸ“Š Cluster {c}:")
    top_players = player_df[player_df['cluster'] == c].nsmallest(10, 'distance_to_center')
    print(top_players[['player', 'position', 'team', 'distance_to_center'] + features])
```

We clusered the players in 4 types. The plot looks quite messy but if we look at the stats it makes completly sense the groups. That is because we are ploting a 14 dimension kmeans in 2D. Now we will plot in 3D to make it easier for the user to see.

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# PCA transformation
pca = PCA(n_components=3)
X_pca_3d = pca.fit_transform(X_scaled)

# Setup color mapping
colors = sns.color_palette('tab10', n_colors=len(set(player_df['cluster'])))
cluster_colors = [colors[c] for c in player_df['cluster']]

# Create figure with 4 views
fig = plt.figure(figsize=(16, 12))

view_angles = [
    (20, 30),   # default
    (60, 20),   # side-top
    (120, 10),  # reverse side
    (300, 50)   # low-angled rotation
]

for i, (azim, elev) in enumerate(view_angles):
    ax = fig.add_subplot(2, 2, i+1, projection='3d')
    ax.scatter(
        X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2],
        c=cluster_colors,
        s=60,
        edgecolor='k',
        alpha=0.8
    )
    ax.set_title(f'3D PCA View (azim={azim}, elev={elev})')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.view_init(elev=elev, azim=azim)

plt.suptitle("Player Clusters from Multiple 3D Angles", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

```

Here the distribution looks better and we can see a separation between the groups

```python
import pandas as pd
import numpy as np

# Compute overall means
overall_means = player_df[features].mean()

# Compute mean of each feature per cluster
cluster_means = player_df.groupby('cluster')[features].mean()

# Get significance by deviation from global mean
significance = (cluster_means - overall_means).abs()

# Display top predictors per cluster
top_n = 15  # top 5 features that most define each cluster
for cluster_id in cluster_means.index:
    print(f"\nðŸ”Ž Cluster {cluster_id} â€” Top {top_n} defining features:")
    top_features = significance.loc[cluster_id].sort_values(ascending=False).head(top_n)
    for feature, diff in top_features.items():
        value = cluster_means.loc[cluster_id, feature]
        print(f"  â€¢ {feature}: {value:.2f} (Î” {diff:.2f})")
```
