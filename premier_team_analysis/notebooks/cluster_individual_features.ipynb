```python
import pandas as pd
from pathlib import Path 


data_dir = r'C:\SportsProject\premier_team_analysis\data'

player_stats_df = pd.read_csv(f'{data_dir}/player_stats.csv')
player_salaries_df = pd.read_csv(f'{data_dir}/player_salaries.csv')
team_stats_df = pd.read_csv(f'{data_dir}/team_stats.csv')
standings_df = pd.read_csv(f'{data_dir}/standings.csv')
fixtures_df = pd.read_csv(f'{data_dir}/fixtures.csv')
team_salary_df = pd.read_csv(f'{data_dir}/team_salary.csv')
player_possession_stats_df = pd.read_csv(f'{data_dir}/player_possession_stats.csv')
team_possession_stats_df = pd.read_csv(f'{data_dir}/team_possession_stats.csv')


#Code from https://www.kaggle.com/code/astonvaz/stats-don-t-lie-premier-league-24-25-deep-dive

df_dict = {
    'team_possession_stats': team_possession_stats_df,
    'player_possession_stats': player_possession_stats_df,
    'team_salary': team_salary_df,
    'standings': standings_df,
    'team_stats': team_stats_df,
    'player_salaries': player_salaries_df,
    'fixtures': fixtures_df,
    'player_stats': player_stats_df
}

def print_df_shape(name, dataframe):
    """Displays dataframe name and shape"""
    print(f"{name} Dataset shape: {dataframe.shape}\n")
    
def display_column_names(name, dataframe):
    """Displays column names in a dataframe"""
    print(f"{name} columns:\n{dataframe.columns.tolist()} \n")
    
def display_first_five_rows(table_name, dataframe):
    print(f"{table_name} data:")
    print(f"{dataframe.head(5)}\n\n")
    
def display_data_types(name, dataframe):
    print(f"{name} data types:")
    print(f"{dataframe.info()}\n")
    
    
for name, dataframe in df_dict.items():
    print_df_shape(name, dataframe)
    display_column_names(name, dataframe)
    display_first_five_rows(name, dataframe)
    display_data_types(name, dataframe)
    
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def expanded_team_performance(df_dict, team_name, season=None):
    """
    Expanded team performance analysis with multiple clustering approaches and additional metrics.
    
    Args:
        df_dict (dict): Dictionary containing all dataframes
        team_name (str): Name of team to analyze
        season (str/int): Optional season filter
    """
    # Merge and prepare data
    fixtures = df_dict['fixtures'].copy()
    team_stats = df_dict['team_stats'].copy()
    team_possession = df_dict['team_possession_stats'].copy()
    standings = df_dict['standings'].copy()
    
    # Filter by season if provided
    if season and 'Date' in fixtures.columns:
        fixtures['Date'] = pd.to_datetime(fixtures['Date'])
        fixtures['season'] = fixtures['Date'].dt.year
        fixtures = fixtures[fixtures['season'] == season]
    
    # Prepare fixtures data
    fixtures['is_home'] = fixtures['Home'] == team_name
    fixtures['opponent'] = np.where(fixtures['is_home'], fixtures['Away'], fixtures['Home'])
    fixtures['team_score'] = np.where(fixtures['is_home'], fixtures['HomeScore'], fixtures['AwayScore'])
    fixtures['opponent_score'] = np.where(fixtures['is_home'], fixtures['AwayScore'], fixtures['HomeScore'])
    fixtures['result'] = np.select(
        [fixtures['team_score'] > fixtures['opponent_score'], 
         fixtures['team_score'] == fixtures['opponent_score']],
        ['win', 'draw'],
        default='loss'
    )
    
    # Merge with possession data
    opponent_possession = team_possession.rename(
        columns={col: f'opp_{col}' for col in team_possession.columns if col != 'team'}
    )
    merged_data = fixtures.merge(
        opponent_possession, 
        left_on='opponent', 
        right_on='team', 
        how='left'
    ).drop(columns=['team'])
    
    # Feature engineering
    merged_data['goal_difference'] = merged_data['team_score'] - merged_data['opponent_score']
    merged_data['is_win'] = (merged_data['result'] == 'win').astype(int)
    merged_data['is_loss'] = (merged_data['result'] == 'loss').astype(int)
    merged_data['points'] = merged_data['result'].map({'win': 3, 'draw': 1, 'loss': 0})
    
## 1. SIMPLE POSSESSION-BASED CLUSTERING
print("\n" + "="*50)
print("SIMPLE POSSESSION-BASED CLUSTERING ANALYSIS")
print("="*50)

possession_data = merged_data[['opp_possession', 'team_score', 'opponent_score']].dropna()

if len(possession_data) >= 3:  # Need at least 3 matches for clustering
    # Simple binary clustering based on opponent possession median
    possession_threshold = possession_data['opp_possession'].median()
    merged_data['possession_cluster'] = np.where(
        merged_data['opp_possession'] > possession_threshold,
        'High possession opponent',
        'Low possession opponent'
    )
    
    # Get match counts per cluster
    match_counts = merged_data['possession_cluster'].value_counts().to_frame('match_count')
    
    # Analyze performance
    possession_perf = merged_data.groupby('possession_cluster').agg({
        'is_win': ['mean', 'sum'],
        'is_loss': ['mean', 'sum'],
        'team_score': 'mean',
        'opponent_score': 'mean',
        'opp_possession': 'mean'
    })
    
    # Combine with match counts
    possession_perf = possession_perf.join(match_counts)
    
    print("\nPerformance by Possession Cluster:")
    print(possession_perf)
    
    # Plot with match counts annotated
    plt.figure(figsize=(12, 6))
    ax = sns.boxplot(
        data=merged_data,
        x='possession_cluster',
        y='goal_difference',
        hue='result',
        palette={'win': 'green', 'draw': 'yellow', 'loss': 'red'}
    )
    
    # Add match count annotations
    for i, cluster in enumerate(possession_perf.index):
        count = possession_perf.loc[cluster, ('match_count', '')]
        ax.text(i, ax.get_ylim()[1]*0.95, 
                f'n={count}', 
                ha='center', va='center',
                bbox=dict(facecolor='white', alpha=0.8))
    
    plt.title(f'Goal Difference by Possession Cluster for {team_name}\n(Median threshold: {possession_threshold:.1f}%)')
    plt.ylabel('Goal Difference')
    plt.xlabel('')
    plt.legend(title='Result')
    plt.tight_layout()
    plt.show()
    
    # Print threshold information
    print(f"\nPossession threshold: {possession_threshold:.1f}% (median opponent possession)")
    print(f"Matches vs high-possession opponents: {match_counts.loc['High possession opponent', 'match_count']}")
    print(f"Matches vs low-possession opponents: {match_counts.loc['Low possession opponent', 'match_count']}")
else:
    print("Not enough data for possession clustering")
    
    ## 2. ADVANCED MATCH TYPE CLUSTERING (Multi-dimensional)
    print("\n" + "="*50)
    print("ADVANCED MATCH TYPE CLUSTERING")
    print("="*50)
    
    cluster_features = [
        'opp_possession', 'opp_touches', 'opp_attacking_touches',
        'team_score', 'opponent_score'
    ]
    cluster_data = merged_data[cluster_features].dropna()
    
    if len(cluster_data) >= 5:  # Need reasonable sample size
        # Standardize
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(cluster_data)
        
        # Determine optimal clusters using silhouette score
        best_k = 2
        best_score = -1
        for k in range(2, min(5, len(cluster_data))):
            kmeans = KMeans(n_clusters=k, random_state=42)
            labels = kmeans.fit_predict(scaled_data)
            score = silhouette_score(scaled_data, labels)
            if score > best_score:
                best_score = score
                best_k = k
        
        # Apply clustering
        kmeans = KMeans(n_clusters=3, random_state=42)
        clusters = kmeans.fit_predict(scaled_data)
        merged_data.loc[cluster_data.index, 'match_cluster'] = clusters
        
        # Analyze clusters
        cluster_summary = merged_data.groupby('match_cluster').agg({
            'opp_possession': ['mean', 'std'],
            'opp_touches': ['mean', 'std'],
            'team_score': ['mean', 'std'],
            'opponent_score': ['mean', 'std'],
            'is_home': ['mean'],
            'is_win': ['mean'],
            'is_loss': ['mean'],
            'points': ['mean', 'count']
        })
        print("\nMatch Cluster Summary:")
        print(cluster_summary)
        
        # Visualize clusters
        plt.figure(figsize=(12, 8))
        sns.scatterplot(
            data=merged_data.dropna(),
            x='opp_possession',
            y='opponent_score',
            hue='match_cluster',
            style='result',
            palette='viridis',
            s=100
        )
        plt.title(f'Advanced Match Clusters for {team_name}')
        plt.xlabel('Opponent Possession (%)')
        plt.ylabel('Goals Conceded')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
    else:
        print("Not enough data for advanced clustering")
    
    
    ## 4. OPPONENT STRENGTH ANALYSIS
    print("\n" + "="*50)
    print("OPPONENT STRENGTH ANALYSIS")
    print("="*50)
    
    if 'standings' in df_dict:
        # Get opponent points per game
        standings = df_dict['standings'].copy()
        standings['ppg'] = standings['points'] / (standings['win'] + standings['loss'] + standings['draw'])
        
        # Merge with fixtures
        opponent_strength = merged_data.merge(
            standings[['team', 'ppg', 'rank']],
            left_on='opponent',
            right_on='team',
            how='left',
            suffixes=('', '_opp')
        )
        
        if not opponent_strength.empty:
            # Categorize opponents
            bins = [0, opponent_strength['ppg'].quantile(0.33), 
                    opponent_strength['ppg'].quantile(0.66), 
                    opponent_strength['ppg'].max()]
            labels = ['Weak', 'Average', 'Strong']
            opponent_strength['opponent_strength'] = pd.cut(
                opponent_strength['ppg'],
                bins=bins,
                labels=labels,
                include_lowest=True
            )
            
            # Analyze performance
            strength_perf = opponent_strength.groupby('opponent_strength').agg({
                'is_win': 'mean',
                'is_loss': 'mean',
                'team_score': 'mean',
                'opponent_score': 'mean',
                'opp_possession': 'mean'
            })
            print("\nPerformance by Opponent Strength:")
            print(strength_perf)
            
            # Plot
            plt.figure(figsize=(10, 6))
            sns.barplot(
                data=opponent_strength,
                x='opponent_strength',
                y='points',
                hue='is_home',
                order=labels,
                ci=None,
                palette=['lightblue', 'darkblue']
            )
            plt.title(f'Average Points Gained by Opponent Strength for {team_name}')
            plt.ylabel('Average Points')
            plt.xlabel('Opponent Strength')
            plt.legend(title='Home/Away', labels=['Away', 'Home'])
            plt.tight_layout()
            plt.show()
    
    ## 5. TIME-BASED PERFORMANCE ANALYSIS
    print("\n" + "="*50)
    print("TIME-BASED PERFORMANCE ANALYSIS")
    print("="*50)
    
    if 'Date' in merged_data.columns:
        merged_data['Date'] = pd.to_datetime(merged_data['Date'])
        merged_data['month'] = merged_data['Date'].dt.month_name()
        merged_data['match_week'] = merged_data.groupby(
            merged_data['Date'].dt.isocalendar().week
        ).ngroup() + 1
        
        # Rolling performance metrics
        merged_data = merged_data.sort_values('Date')
        merged_data['rolling_points'] = merged_data['points'].rolling(5, min_periods=1).mean()
        merged_data['rolling_goals'] = merged_data['team_score'].rolling(5, min_periods=1).mean()
        merged_data['rolling_conceded'] = merged_data['opponent_score'].rolling(5, min_periods=1).mean()
        
        # Plot rolling performance
        plt.figure(figsize=(12, 6))
        plt.plot(merged_data['match_week'], merged_data['rolling_points'], label='5-game avg points')
        plt.plot(merged_data['match_week'], merged_data['rolling_goals'], label='5-game avg goals scored')
        plt.plot(merged_data['match_week'], merged_data['rolling_conceded'], label='5-game avg goals conceded')
        plt.xlabel('Match Week')
        plt.title(f'Rolling 5-Game Performance for {team_name}')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        
        # Monthly performance
        monthly_perf = merged_data.groupby('month').agg({
            'points': 'mean',
            'team_score': 'mean',
            'opponent_score': 'mean'
        })
        print("\nMonthly Performance:")
        print(monthly_perf)
    
    return merged_data

# Example usage:
expanded_results = expanded_team_performance(df_dict, "Liverpool")
```
